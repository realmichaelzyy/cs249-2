## Problem 1.  What <i>kind</i> of distribution does the number of Beneficiaries appear to follow?

```python
%%R -w 600 -h 400

NumberOfBeneficiaries = NumericData[,2]

# intuitively, it would make sense if this number followed some kind of arrival distribution

# The number of beneficiaries is very large in some cases (see the summary above),
# so let's try a plot of log10(number of beneficiaries)
# to see how many there are

hist( log10(NumberOfBeneficiaries),  col="deepskyblue", breaks=50 )

```

```python
%%R
# The histogram above shows that most numbers of beneficiaries are below 10^(3.7) = 5,000.

# The R expression  (NumberOfBeneficiaries < 5000)  yields a boolean vector.
# It can be used as a subscript to select all entries of a vector satisfying this boolean:


NumberOfBeneficiariesBelow5K = NumberOfBeneficiaries[ (NumberOfBeneficiaries < 5000) ]

hist( NumberOfBeneficiariesBelow5K,  col="deepskyblue", border=NA, breaks=500)
```

```python
%%R
# Try to fit a lognormal to this dataset....

library(MASS)

NumberOfBeneficiariesBelow5K = NumberOfBeneficiaries[ (NumberOfBeneficiaries < 5000) ]

hist( NumberOfBeneficiariesBelow5K,  col="deepskyblue", border=NA, breaks=500,prob=TRUE  )

params = fitdistr(NumberOfBeneficiariesBelow5K,"lognormal")

curve(dlnorm(x,meanlog=params$estimate[1],sdlog=params$estimate[2]),col="red",add=TRUE)

mtext( "a lognormal fit seems OK", side=3, col="red" )
```

## Problem 2.  Fit a distribution to:  log10( submitted total charge amounts )

```python
%%R

hist( log10(SubmittedChargeAmounts), probability=TRUE, border=NA, col="palegreen", breaks=500 )

parameters = fitdistr( log10(SubmittedChargeAmounts), "lognormal" )

curve( dlnorm(x, meanlog=parameters$estimate[1], sdlog=parameters$estimate[2]), col="red", add=TRUE )  # doesn't fit very well

mtext( "rats, that's annoying, the lognormal distribution doesn't fit very well", side=3, col="red" )

print(parameters$loglik)
```

```python
%%R
hist( log10(SubmittedChargeAmounts), probability=TRUE, border=NA, col="palegreen", breaks=500 )

parameters = fitdistr( log10(SubmittedChargeAmounts), "lognormal" )
curve( dlnorm(x, meanlog=parameters$estimate[1], sdlog=parameters$estimate[2]), col="red", add=TRUE )  # doesn't fit very well
mtext( "that's annoying, the lognormal distribution doesn't fit very well", side=3, col="red" )

more_parameters = fitdistr( log10(SubmittedChargeAmounts), "gamma" )
curve( dgamma(x, shape=more_parameters$estimate[1], rate=more_parameters$estimate[2]), col="blue", add=TRUE )  # doesn't fit very well
mtext( "good grief, the gamma distribution doesn't fit well either", line=-1, side=3, col="blue" )

hist( log10(SubmittedChargeAmounts), probability=TRUE, border=NA, col="palegreen", breaks=500 )

parameters = fitdistr( log10(SubmittedChargeAmounts), "weibull" )

curve( dweibull(x, shape=parameters$estimate[1], scale=parameters$estimate[2]), col="red", add=TRUE )

mtext( "the weibull distribution fits a little better", side=3, col="red" )

print(parameters$loglik)
```

## Problem 3.  Fit a distribution to: log10( average charge amounts )

```python
%%R

AverageChargeAmounts =  log10( SubmittedChargeAmounts / NumberOfBeneficiaries )


average_amounts_histogram = hist( AverageChargeAmounts, 
                                 probability=TRUE, border=NA, col="cyan", breaks=500 )

mtext("this looks like a mixture, but you can ignore the amounts below 1.7", side=3, col="blue")

AverageChargeAmounts = AverageChargeAmounts[ AverageChargeAmounts >= 1.7 ]

# what is a good fit for the distribution of AverageChargeAmounts? (ignoring the bump at the low end)

# try using fitdistr() to approximate a fit

# fit the log of the data to a normal distribution to test for log-normal distribution fit
normalparams = fitdistr(AverageChargeAmounts,"normal")
curve( dnorm(x, mean=normalparams$estimate[1], sd=normalparams$estimate[2]), col="red", add=TRUE )

gammaparams = fitdistr(AverageChargeAmounts,"gamma")
curve( dgamma(x, shape=gammaparams$estimate[1], rate=gammaparams$estimate[2]), scale=gammaparams$estimate[3], col="blue", add=TRUE )

cauchyparams = fitdistr(AverageChargeAmounts,"cauchy")
curve( dcauchy(x, location=cauchyparams$estimate[1], scale=cauchyparams$estimate[2]), col="dark red", add=TRUE )

cat(sprintf("normal loglik: %0.2f, gamma loglik: %0.2f, cauchy loglik: %0.2f", normalparams$loglik, gammaparams$loglik, cauchyparams$loglik))
```

### The normal distribution looks like it has the best fit, although it's log likelihood is slightly worse than gamma. I think this is probably because the gamma distribution is skewed slightly to the left, and so it picks up more of the left mixture "hump". The normal seems to be the best fit though.

## Problem 5.  Does Total Charge Amount by State follow a Power Law?

```python
%%R -w 800 -h 800

total_charge_by_state = aggregate( total_submitted_chrg_amt  ~  nppes_provider_state,  data=MedicarePaymentData,  sum  )

hist( log( total_charge_by_state[,2] ), col="seagreen", breaks=20 )

ranked_by_state = total_charge_by_state[ order( total_charge_by_state[,2], decreasing=TRUE ), ]

# keep only the top 50 states (ignore international data)
ranked_by_state = ranked_by_state[1:50,]

ranks = 1:nrow(ranked_by_state)

plot(   log(ranks), log(ranked_by_state$total_submitted_chrg_amt), type="l", col="blue" )  # line segments first
points( log(ranks), log(ranked_by_state$total_submitted_chrg_amt), col="darkblue", pch=20 )  # add dots

text( log(ranks), log(ranked_by_state$total_submitted_chrg_amt),
     ranked_by_state$nppes_provider_state, pos=2, cex=0.75, col="red" )

# ... is this relationship a power law?  Is the plot essentially log-log-linear?
# (Does it look like "the rich get richer" ?)
```

### It does not look log-log-linear to me, so I don't think it is a power law distribution.

```python
%%R -w 800 -h 400

total_charge_by_state = aggregate( total_submitted_chrg_amt  ~  nppes_provider_state,  data=MedicarePaymentData,  sum  )

hist(total_charge_by_state[,2] , col="seagreen", breaks=20, prob=TRUE )

params = fitdistr(total_charge_by_state[,2] ,"geometric")
curve(dgeom(x,prob=params$estimate[1]),col="red",add=TRUE)

params = fitdistr(total_charge_by_state[,2] ,"exponential")
curve(dexp(x,rate=params$estimate[1]),col="blue",add=TRUE)
```

### Tried fitting geometric and exponential distributions, and neither one looks like a good fit...supporting our previous analysis that thought it did not look log-log-linear and therefore it didn't follow a power-law distribution

## Problem 5.  What is the distribution of maximum charges made by the different kinds of Providers?

```python
%%R -w 1100

max_charge_by_provider_type = aggregate( total_submitted_chrg_amt  ~ provider_type,  data=MedicarePaymentData,  max  )

hist( log10( max_charge_by_provider_type$total_submitted_chrg_amt ), col="purple", probability=TRUE )

rank_ordering_by_max_charge = order( log10(max_charge_by_provider_type[,2]), decreasing=TRUE )

ranked_by_type = max_charge_by_provider_type[ rank_ordering_by_max_charge , ]

print(ranked_by_type)

ranks = 1:nrow(ranked_by_type)

plot(   log(ranks), log(ranked_by_type$total_submitted_chrg_amt), type="l", col="blue" )
points( log(ranks), log(ranked_by_type$total_submitted_chrg_amt), col="darkblue", pch=20 )

text( log(ranks), log(ranked_by_type$total_submitted_chrg_amt),
     ranked_by_type$provider_type, pos=2, cex=0.5, col="red" )
```

### Definitely not log-log linear, so not a power law distribution. Let's try to find something that does fit it better...

```python
%%R -w 1100

hist( log10( max_charge_by_provider_type$total_submitted_chrg_amt ), col="purple", probability=TRUE, breaks=10 )

# fit the log of the data to a normal distribution
params = fitdistr(log10( max_charge_by_provider_type$total_submitted_chrg_amt ), "normal")
curve( dnorm(x, mean=params$estimate[1], sd=params$estimate[2]), col="red", add=TRUE )
```

### It seems to fit a log-normal distribution fairly well, but it's hard to tell without more data.
