# The Supervised Learning paradigm in R

## Example:  Supervised learning on the Iris dataset

```python
%%R

data(iris)
print(str(iris))

n <- dim(iris)[1]
p <- dim(iris)[2]

print(table(iris$Species))

#------------------------------------------------------------------------
#   generate a training set, and create a model on the training set
#------------------------------------------------------------------------

iris_ids <- (1:n)

training_set <- sample(iris_ids, 75)     # Generate a random sample of size 75
print(training_set)

testing_set  <- iris_ids[-training_set]  # The set complement of training_set
print(testing_set)

print(table(iris$Species[training_set]))    # Tabulate species for the training set
print(table(iris$Species[testing_set]))     # Tabulate species for the testing set

#  plot the data as a parallel coordinates plot  (parcoord is in the MASS package)

parcoord(iris[,1:4],  col=c("red", "green3", "blue")[unclass(iris$Species)])
```

## fit an LDA model to the training set data

```python
%%R

#  lda is in the MASS package

ldaModel <- lda( Species ~ Petal.Length + Petal.Width + Sepal.Length + Sepal.Width,
                 data = iris, subset = training_set )

# Equivalent abbreviation:

ldaModel <- lda( Species ~ .,  data = iris, subset = training_set )


print(ldaModel)
```

## Evaluate how well the LDA model predicts classes for the remaining data (i.e., the testing_set)

```python
%%R

# Make class predictions on the test dataset using our LDA model
#  and then check the predicted class results against the true class values.

predictions <- predict( ldaModel, iris[ testing_set ,] )

print(predictions$class)

incorrect.prediction  <-  (predictions$class != iris$Species[testing_set])

print(testing_set[ incorrect.prediction ])          # ids  for which prediction was incorrect

print(iris[ testing_set[incorrect.prediction], ])   # data for which prediction was incorrect

confusion_matrix <- table( iris$Species[testing_set], predictions$class )
print(confusion_matrix)

m <- length(testing_set)
print(m)

accuracy <- (m - sum(incorrect.prediction))/m
print(accuracy)
```

## Never mix the training set with the testing set !!

```python
%%R

#  out of curiousity, let us see how well the model works on the training set.
#  (It is a terrible idea to do this in general)

predictions <- predict( ldaModel, iris[ training_set, ] )
print(predictions$class)

incorrect.prediction  <-  (predictions$class != iris$Species[training_set] )

print(iris_ids[ incorrect.prediction ]) # ids  for which prediction was incorrect
print(iris[ incorrect.prediction , ])   # data for which prediction was incorrect

accuracy <- (n - sum(incorrect.prediction))/n
print(accuracy)

confusion_matrix <- table( iris$Species[training_set], predictions$class )
print(confusion_matrix)
```

## Training another kind of model:  the SVM

```python
%%R

# Now train an SVM model, and evaluate its accuracy

library(e1071)

svmModel <- svm( Species ~ ., data=iris, subset = training_set )
print(summary(svmModel))

svm.predictions <- predict( svmModel, iris[ testing_set ,] )

svm.incorrect.prediction  <-  (svm.predictions != iris$Species[testing_set])
print(testing_set[svm.incorrect.prediction])

svm.confusion.matrix <- table( iris$Species[testing_set], svm.predictions )
print(svm.confusion.matrix)

svm.accuracy <- (m - length(training_set[svm.incorrect.prediction])) / m
print(svm.accuracy)
```

# Problem 3.  Do problem 10 in [ISL] Chapter 4, for the Weekly dataset.

```python
%%R -w 900 -h 900

data(Weekly)

print(summary(Weekly))

# a

pairs(Weekly)

print(cor(Weekly[,c(-1,-1,-9)]))

```

```python
%%R
# c

logRegmodel = glm(Direction~Lag1+Lag2+Lag3+Lag4+Lag5+Volume,data=Weekly,family=binomial)
print(summary(logRegmodel))

# lag2 has a significant p-value, although at 0.03 it isn't super significant.

predictions <- predict( logRegmodel, Weekly )
#print(predictions$class)
downInd = predictions <= 0
predictions[predictions > 0] = "Up"
predictions[downInd] = "Down"

incorrect.prediction  <-  (predictions != Weekly$Direction )

#print(Weekly[ incorrect.prediction ])          # ids  for which prediction was incorrect
#print(Weekly[ Weekly[incorrect.prediction], ])   # data for which prediction was incorrect

cat(sprintf("Logistic Regression Proportion correct: %0.3f\n\nConfusion matrix: \n", 1-sum(incorrect.prediction)/length(incorrect.prediction)))

confusion_matrix <- table( Weekly$Direction, predictions)
print(confusion_matrix)

# The confusion matrix shows that the model is better at predicting up than it is at predicting down, because the vast 
# majority of predictions are "Up" while the actual market data is not.
```

```python
%%R
# d

trainingIds = Weekly$Year <= 2008
testingIds = Weekly$Year > 2008

logRegmodel = glm(Direction~Lag2,data=Weekly[trainingIds,],family=binomial)
#print(summary(logRegmodel))

# lag2 has a significant p-value, although at 0.03 it isn't super significant.

predictions <- predict( logRegmodel, Weekly[testingIds,] )

downInd = predictions <= 0
predictions[predictions > 0] = "Up"
predictions[downInd] = "Down"

incorrect.prediction  <-  (predictions != Weekly$Direction[testingIds] )

cat(sprintf("Proportion correct: %0.3f\n\nConfusion matrix: \n", 1-sum(incorrect.prediction)/length(incorrect.prediction)))

confusion_matrix <- table( Weekly$Direction[testingIds], predictions)
print(confusion_matrix)
```

```python
%%R

# e

ldaModel <- lda(Direction ~ Lag2,  data = Weekly, subset = trainingIds )
#logRegmodel = glm(Direction~Lag2,data=Weekly[trainingIds,],family=binomial)

predictions <- predict( ldaModel, Weekly[ testingIds ,] )

#print(predictions$class)

incorrect.prediction  <-  (predictions$class != Weekly$Direction[testingIds])

cat(sprintf("LDA Proportion correct: %0.3f\n\nConfusion matrix: \n", 1-sum(incorrect.prediction)/length(incorrect.prediction)))

confusion_matrix <- table( Weekly$Direction[testingIds], predictions$class )
print(confusion_matrix)
```

```python
%%R

# f

qdaModel <- qda(Direction ~ Lag2,  data = Weekly, subset = trainingIds )
#logRegmodel = glm(Direction~Lag2,data=Weekly[trainingIds,],family=binomial)

predictions <- predict( qdaModel, Weekly[ testingIds ,] )

incorrect.prediction  <-  (predictions$class != Weekly$Direction[testingIds])

cat(sprintf("QDA Proportion correct: %0.3f\n\nConfusion matrix: \n", 1-sum(incorrect.prediction)/length(incorrect.prediction)))

confusion_matrix <- table( Weekly$Direction[testingIds], predictions$class )
print(confusion_matrix)
```

```python
%%R

# g

library(class)

trainX = cbind(Weekly$Lag2[trainingIds])
trainY = Weekly$Direction[trainingIds]
testX = cbind(Weekly$Lag2[testingIds])

predictions = knn(trainX,testX,trainY,k=1)

incorrect.prediction  <-  (predictions != Weekly$Direction[testingIds])

cat(sprintf("Knn Proportion correct: %0.3f\n\nConfusion matrix: \n", 1-sum(incorrect.prediction)/length(incorrect.prediction)))

confusion_matrix <- table( Weekly$Direction[testingIds], predictions )
print(confusion_matrix)
```

```python
%%R

# h

# LDA and Logistic Regression have the best performance, with 62.5% accuracy

# i

# LDA experiment
ldaModel <- lda(Direction ~ Lag1*Lag2+Lag3,  data = Weekly, subset = trainingIds )
predictions <- predict( ldaModel, Weekly[ testingIds ,] )
incorrect.prediction  <-  (predictions$class != Weekly$Direction[testingIds])
cat(sprintf("LDS Proportion correct: %0.3f\n", 1-sum(incorrect.prediction)/length(incorrect.prediction)))

# QDA experiment
qdaModel <- qda(Direction ~ Lag1+I(Lag1^2)+I(Lag1^3),  data = Weekly, subset = trainingIds )
predictions <- predict( qdaModel, Weekly[ testingIds ,] )
incorrect.prediction  <-  (predictions$class != Weekly$Direction[testingIds])
cat(sprintf("QDA Proportion correct: %0.3f\n", 1-sum(incorrect.prediction)/length(incorrect.prediction)))

# Knn experiment
trainX = Weekly[trainingIds,c(1,2)]
trainY = Weekly[trainingIds,9]
testX = Weekly[testingIds,c(1,2)] #cbind(Weekly$Lag2[testingIds])
predictions = knn(trainX,testX,trainY,k=1)
incorrect.prediction  <-  (predictions != Weekly$Direction[testingIds])
cat(sprintf("Knn Proportion correct: %0.3f\n", 1-sum(incorrect.prediction)/length(incorrect.prediction)))

# can't really get better
```

# Problem 4.  Develop LDA, QDA, SVM, and logistic regression models for the SPAM dataset (in the ElemStatLearn library)

```python
%%R

data(spam)

#print(summary(spam))

# ........ (note section 4.4.5 of [ESL] compares LDA and logistic regression)


n <- dim(spam)[1]
p <- dim(spam)[2]

#------------------------------------------------------------------------
#   generate a training set, and create a model on the training set
#------------------------------------------------------------------------

total_ids <- (1:n)
training_set <- sample(total_ids, round(n*0.7))     # Generate a random sample of size 0.75*n
testing_set  <- total_ids[-training_set]  # The set complement of training_set

print(table(spam$spam[training_set]))    # Tabulate species for the training set
print(table(spam$spam[testing_set]))     # Tabulate species for the testing set

# LDA model
ldaModel <- lda( spam ~ .,  data = spam, subset = training_set )

predictions <- predict( ldaModel, spam[ testing_set ,] )
incorrect.prediction  <-  (predictions$class != spam$spam[testing_set])
cat(sprintf("\nLDA Proportion correct: %0.3f\nConfusion matrix: \n", 1-sum(incorrect.prediction)/length(incorrect.prediction)))
confusion_matrix <- table( spam$spam[testing_set], predictions$class )
print(confusion_matrix)

# QDA model
qdaModel <- qda( spam ~ .,  data = spam, subset = training_set )

predictions <- predict( qdaModel, spam[ testing_set ,] )
incorrect.prediction  <-  (predictions$class != spam$spam[testing_set])
cat(sprintf("\nQDA Proportion correct: %0.3f\nConfusion matrix: \n", 1-sum(incorrect.prediction)/length(incorrect.prediction)))
confusion_matrix <- table( spam$spam[testing_set], predictions$class )
print(confusion_matrix)

# SVM model
svmModel <- svm( spam ~ .,  data = spam, subset = training_set )

predictions <- predict( svmModel, spam[ testing_set ,] )
incorrect.prediction  <-  (predictions != spam$spam[testing_set])
cat(sprintf("\nSVM Proportion correct: %0.3f\nConfusion matrix: \n", 1-sum(incorrect.prediction)/length(incorrect.prediction)))
confusion_matrix <- table( spam$spam[testing_set], predictions )
print(confusion_matrix)

# Logistic regression model
logRegmodel = glm(spam~.,data=spam[training_set,],family=binomial)

predictions <- predict( logRegmodel, spam[testing_set,] )

downInd = predictions <= 0
predictions[predictions > 0] = "spam"
predictions[downInd] = "email"

incorrect.prediction  <-  (predictions != spam$spam[testing_set] )

cat(sprintf("\nLogistic Regression Proportion correct: %0.3f\nConfusion matrix: \n", 1-sum(incorrect.prediction)/length(incorrect.prediction)))

confusion_matrix <- table( spam$spam[testing_set], predictions)
print(confusion_matrix)
```
